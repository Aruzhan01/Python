```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color

# Load a sample image from skimage
#image = data.astronaut()
image = cv2.imread('cv.png')

# Convert the image to blue and white
# We will zero out the red and green channels
blue_white_image = np.copy(image)
blue_white_image[:, :, 0] = 0  # Red channel
blue_white_image[:, :, 1] = 0  # Green channel

# Enhance white: for simplicity, let's increase brightness of the blue channel
# We identify bright areas and make them brighter
bright_areas = blue_white_image[:, :, 2] > 120
blue_white_image[bright_areas, 2] = 255

# Display the original and the modified image
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(blue_white_image)
plt.title('Blue and White Image')
plt.axis('off')

plt.show()
```
